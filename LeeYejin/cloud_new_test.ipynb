```python
# @title 메타데이터 생성
from sklearn.model_selection import train_test_split

def load_data(base_dir='test_data'):
  images_path = []
  labels = []
  class_names = []

  train_base = os.path.join(base_dir,"test_cloud_img")

  for class_name in os.listdir(train_base):
    class_name_path = os.path.join(train_base, class_name)

    # 디렉토리인 경우만
    if os.path.isdir(class_name_path):
      for image_name in os.listdir(class_name_path):
        image_path = os.path.join(class_name_path, image_name)
        images_path.append(image_path)
        labels.append(class_name)

  return np.array(images_path), np.array(labels)

images_path, labels = load_data()
print(images_path.shape, labels.shape)
print(images_path, labels)
type(images_path), type(labels)
```

```python
batch_images=[]
for img_path in images_path:

  image = cv2.imread(img_path)

  image = cv2.resize(image, (224,224))
  image = np.array(image)
  batch_images.append(image)
```

```python
print(batch_images)
```

```python
import cv2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

class_names = [
        'Ac',
        'As',
        'Cb',
        'Cc',
        'Ci',
        'Cs',
        'Ct',
        'Cu',
        'Ns',
        'Sc',
        'St'
    ]


IMAGE_SIZE = 224
for i,a_image in enumerate(batch_images):
  a_image = preprocess_input(a_image)
  batch_image = a_image.reshape(1, IMAGE_SIZE, IMAGE_SIZE, 3)
  pred_proba = model.predict(batch_image)
  pred = np.argmax(pred_proba)
  pred_label = class_names[pred]
  print(pred_proba)
  print(f'예측 라벨 : {pred_label},실제 label :{labels[i]}')
  print(f'예측 확률 : {pred_proba[0][pred]:.4f}')
```
